# ***using Spacy***

```python
import spacy
nlp = spacy.load("en_core_web_sm")
from spacy.lang.en.stop_words import STOP_WORDS
from spacy import displacy
```

```python
from google.colab import drive
drive.mount('/content/drive' , force_remount=True)
dis_summary_df = pd.read_csv('/content/drive/MyDrive/MIMIC3FLDS/discharge_summary_arf_n.csv.gz', index_col=0)
dis_summary_df
```

```python
#tokenize and cleans text
def extract_tokenize_spacy(text):
    doc = nlp(text)
    tokens = [token.text.lower() for token in doc if (token.is_alpha or token.is_digit) and not token.is_punct]
    tokens = [token for token in tokens if token not in STOP_WORDS]

    processes_text = ' '.join(tokens)
    return processes_text

def extract_entities_spacy(text):
    doc = nlp(text)
    for ent in doc.ents:
        print(ent.text, ent.start_char, ent.end_char, ent.label_)
        print("************************************************************")
    return doc



```

```python
dis_summary_subset_df = dis_summary_df.head(10)
```

```python
dis_summary_subset_df ['PROCESSED_TEXT'] = dis_summary_subset_df['TEXT'].apply(extract_tokenize_spacy)
dis_summary_subset_df
```

```python
extract_entities_spacy(dis_summary_subset_df.iloc[0]['PROCESSED_TEXT'])
```

```python
displacy.render(nlp(dis_summary_subset_df.iloc[0]['PROCESSED_TEXT']), style="ent",jupyter=True)
```
