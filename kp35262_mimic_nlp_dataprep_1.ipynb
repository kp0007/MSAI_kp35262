```python
noteevents_df=pd.read_csv('NOTEEVENTS.csv',low_memory=False).set_index('ROW_ID')
```

```python
diagnoses_icd_df=pd.read_csv('DIAGNOSES_ICD.csv').set_index('ROW_ID')
```

```python
discharge_df=noteevents_df.loc[noteevents_df['CATEGORY']=='Discharge summary',['SUBJECT_ID','HADM_ID','TEXT']]
discharge_df.head()
```

```python
disease_icd = ['51881']
disease_df = diagnoses_icd_df[diagnoses_icd_df['ICD9_CODE'].isin(disease_icd)].copy()
```

```python
disease_type = {'51881': 'Acute respiratry failure'}
disease_df['DISEASE_TYPE'] = disease_df['ICD9_CODE'].map(disease_type)
```

```python
discharge_disease_df = pd.merge(discharge_df,disease_df, on=['SUBJECT_ID', 'HADM_ID'] , how='inner')
discharge_disease_df.head()
discharge_disease_df.to_csv('/content/drive/MyDrive/MIMIC3FLDS/discharge_summary_arf_n.csv.gz')
```
